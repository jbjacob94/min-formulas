```python
# import python packages
import os
import numpy as np
import MinFormula as mf
import pandas as pd
from pathlib import Path
```

```python
# load file, check data
filenames= os.listdir('Compo_test')
filepaths = [os.path.join('Compo_test', f) for f in filenames]

#minerals = [Path(i).stem for i in filenames]
DATA = [pd.read_csv (i) for i in filepaths]
element_list = ["SiO2","TiO2","Al2O3","Cr2O3","FeO","MnO","MgO","CaO","Na2O","K2O","P2O5"] # do not change!
print(DATA)
```

```python
# recalculate garnet formulas
# Extract composition and convert dataframe to a numpy array
data = DATA[0]
compo_grt = data[element_list].to_numpy()
compo_grt=np.nan_to_num(compo_grt, copy=False, nan=0.0)

# StrGarnet function
Si,Ti,Al,Cr,Fe,Mn,Mg,Ca,SumCat,Si_T,Al_T,Al_M2,Cr_M2,Ti_M2,Fe_M1,Mg_M1,Ca_M1,Mn_M1,SumM2,SumM1,Xalm,Xpyr,Xgrs,Xsps = np.round(mf.StrGarnet(compo_grt),2)

# create new data frame with structural formulas
col_names = ['Si','Ti','Al','Cr','Fe','Mn','Mg','Ca','SumCat','Xalm','Xpyr','Xgrs','Xsps']
strform = np.transpose(np.array([Si,Ti,Al,Cr,Fe,Mn,Mg,Ca,SumCat,Xalm,Xpyr,Xgrs,Xsps]))
df = pd.DataFrame(data=strform,columns=col_names)

# concatenate to main dataframe
data=pd.concat([data, df], axis=1)
print(data)

# export to csv file
#data.to_csv('garnet_recalc.csv',sep=",", na_rep='na')
```
